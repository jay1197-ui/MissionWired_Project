```python
# Importing Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading constituent information data
cons_df=pd.read_csv("https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons.csv")
cons_df.head()
```

```python
# Displaying the columns of the cons_df DataFrame
cons_df.columns
```

```python
# Loading email addresses data
cons_email_df=pd.read_csv("https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email.csv")
cons_email_df.head()
```

```python
# Displaying the columns of the cons_email_df DataFrame
cons_email_df.columns
```

```python
# Checking for missing values in the cons_email_df DataFrame
cons_email_df.isnull().sum()
```

```python
# Loading subscription status data
subscription_df=pd.read_csv("https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email_chapter_subscription.csv")
subscription_df.head()
```

```python
# Displaying the columns of the subscription_df DataFrame
subscription_df.columns
```

```python
# As per requirement Filtering subscription_df to keep only the rows where chapter_id is 1
subscription_df= subscription_df[subscription_df['chapter_id']==1]
```

```python
# Merging constituent data with their email addresses using 'cons_id'
# This will create a dataset that includes both personal details and email addresses
merged_df=pd.merge(cons_df,cons_email_df,on='cons_id',how='left')
merged_df.head()
```

```python
# Displaying the columns of the merged_df DataFrame
merged_df.columns
```

```python
# Checking for missing values in the merged_df DataFrame
merged_df.isnull().sum()
```

```python
# Filter the merged dataframe to only include rows where 'is_primary' is 1.
# This ensures we are working with the constituent's primary email address.
primary_df=merged_df[merged_df['is_primary']==1]
primary_df.head()
```

```python
# Checking for missing values in the primary_df DataFrame
primary_df.isnull().sum()
```

```python
# we are using final_df to see if the constiuent is still subscribed or unsubscribed
final_df=pd.merge(primary_df,subscription_df,on='cons_email_id',how='left')
final_df.head()
```

```python
# Checking for missing values in the final_df DataFrame
final_df.isnull().sum()
```

```python
# Identifying unique chapter_id values for rows with missing email addresses
final_df[final_df['email'].isnull()]['chapter_id'].unique()
```

```python
# Getting unique chapter_id values from the final_df DataFrame
final_df['chapter_id'].unique()
```

```python
# Checking for missing values in the final_df DataFrame
final_df.columns
```

```python
# Fill missing values in the 'isunsub' column with 0, assuming the constituent is still subscribed.
final_df['isunsub'] = final_df['isunsub'].fillna(0)
# Fill missing values in the 'source' column with 'unknown', indicating the acquisition source is not available.
final_df['source'] = final_df['source'].fillna('unknown')
final_df.head()
```

```python
print(final_df['isunsub'].isnull().sum())
```

```python
final_df['isunsub'].unique()
```

```python
# Creating the 'people_df' dataframe with the required columns as per project specifications
#Here we are considering create_dt_x and modified_dt_x because our main goal is to involve all the constituent information
# As per requirement we need person creation datetime and person updated datetime
people_df=pd.DataFrame({
'email' : final_df['email'], # Email address of the person
'code' : final_df['source'], # Source code indicating acquisition source
'is_unsub' : final_df['isunsub'].astype(int), # Unsubscribe status (1 for unsubscribed, 0 for subscribed)
'created_dt' : pd.to_datetime(final_df['create_dt_x']), #Person creation datetime
'updated_dt' : pd.to_datetime(final_df['modified_dt_x']) #person updated datetime

})
```

```python
# Exporting the 'people_df' dataframe to a CSV file named 'people.csv'
# 'index=False' ensures that row indices are not included in the CSV
# 'header=True' ensures that the column names are included in the CSV file
people_df.to_csv('people.csv',index=False,header=True)
```

```python
people_df.head()
```

```python
#Create Acquisition facts file
# Group the 'people_df' dataframe by the 'created_dt' date and count the number of acquisitions per day
acquisition_facts_df= people_df.groupby(people_df['created_dt'].dt.date).size().reset_index(name='acquisitions')
# Rename the columns of the resulting dataframe for clarity
acquisition_facts_df.columns = ['acquisition_date', 'acquisitions']
acquisition_facts_df
```

```python
# Exporting the 'acquisition_facts_df' dataframe to a CSV file named 'acquisition_facts.csv'
# 'index=False' ensures that row indices are not included in the CSV
acquisition_facts_df.to_csv('acquisition_facts.csv', index=False)
```

```python
# Visualizing the distribution of daily acquisitions
# Create a histogram
plt.figure(figsize=(12, 6))
sns.histplot(acquisition_facts_df['acquisitions'], kde=True, bins=30)

# Set labels and title
plt.xlabel('Number of Acquisitions')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Acquisitions')

# Show the plot
plt.tight_layout()
plt.show()
```
